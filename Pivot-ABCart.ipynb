```python
import pandas as pd

df = pd.read_csv('D:/DYD/AI-ML/Python-Scripts/Data/ABCart.csv')
df.head()

print("Sample Data:")
print(df.head())

pivot_table = pd.pivot_table(
    df,
    values='mobile',        # Column to aggregate
    index='CartCreationDate',      # Row index
    columns='Package',      # Column index
    aggfunc='count',         # Aggregation function ('sum', 'count', 'mean', etc.)
    fill_value=0           # Replace NaNs with 0
)

print("\nPivot Table:")
print(pivot_table)

pivot_table.to_csv('D:/DYD/AI-ML/Python-Scripts/Data/pivot_output.csv')
```
