```python
import pandas as pd

# Step 1: Load CSV
df = pd.read_csv('D:/DYD/AI-ML/Python-Scripts/Data/ABCart.csv')

# Step 2: Convert CartCreationDate to datetime
df['CartCreationDate'] = pd.to_datetime(df['CartCreationDate'], errors='coerce')

# Optional: Show unique date values or range
print("Available date range:", df['CartCreationDate'].min(), "to", df['CartCreationDate'].max())

# Step 3: Filter by date range (modify start_date and end_date as needed)
start_date = '2025-05-01'
end_date = '2025-05-17'

filtered_df = df[(df['CartCreationDate'] >= start_date) & (df['CartCreationDate'] <= end_date)]

print(f"\nFiltered data from {start_date} to {end_date}:")
print(filtered_df.head())

# Step 4: Create pivot table
pivot_table = pd.pivot_table(
    filtered_df,
    values='mobile',
    index='CartCreationDate',
    columns='Package',
    aggfunc='count',
    fill_value=0
)

print("\nPivot Table (Filtered):")
print(pivot_table)

# Step 5: Save output
pivot_table.to_csv('D:/DYD/AI-ML/Python-Scripts/Data/pivot_output_filtered.csv')
```
